## Movie Inventory 



```python
import pandas as pd
customer_df = pd.read_csv(r'C:\data\customer.csv')
customer_df.head()
```

```python
import pandas as pd
inventory_df = pd.read_csv(r'C:\data\inventory.csv')
inventory_df.head()
```

```python
import pandas as pd
payment_df = pd.read_csv(r'C:\data\payment.csv')
payment_df.head()
```

```python
import pandas as pd
rental_df = pd.read_csv(r'C:\data\rental.csv')
rental_df.head()
```

```python
import pandas as pd
staff_df = pd.read_csv(r'C:\data\staff.csv')
staff_df.head()
```

```python
import pandas as pd
store_df = pd.read_csv(r'C:\data\store.csv')
store_df.head()
```

```python
import pandas as pd
address_df = pd.read_csv(r'C:\data\address.csv')
address_df.head()
```

```python
import pandas as pd
film_df = pd.read_csv(r'C:\data\film.csv')
film_df.head()
```

```python
# Top 5, sum amount of payment grouped by customer_id
df = payment_df.groupby(['customer_id']).sum()
df.sort_values(['amount'], ascending=[0]).head()
```

```python
# Top 5, list of rentals from the rental data set that were rented in 2005
df = rental_df[(rental_df['rental_date']>= '2005-01-01') & (rental_df['rental_date']<= '2005-12-31')].sort_values(['return_date'], ascending=[1])
df.head()
```

```python
# Rental count from rental data set grouped by staff_id
df = rental_df.groupby(['staff_id']).count()
df.head()[['rental_id']]
```

```python
df.merge(staff_df, on=['staff_id'], how='left')[['first_name','last_name','rental_id']]
```

```python
# join with inventory, store, and address data frames to get store address
# join with customer data frame to get customer first and last name,
# join with staff data frame to get customer first and last name,
# # join with film  data set to get film name and description
rental_df.merge(inventory_df, on=['inventory_id'],how='left').merge(store_df,on = ['store_id'],how='left').merge(address_df, on = ['address_id'], how = 'left').merge(customer_df, on =['customer_id']).merge(staff_df, on = ['staff_id'],how='left').merge(film_df, on=['film_id'],how = 'left')[['rental_date','return_date','first_name_x','last_name_x','title','description','address','first_name_y','last_name_y']].sort_values(['return_date'], ascending=[1]).head()
```

```python
# join with staff data set to get names
df[['rental_id']].merge(staff_df,on=['staff_id'],how='inner')[['first_name','last_name','amount']]
```

```python
# Top 5 list of customer_id whose sum amount of payment in 2005 is more than $200
df = payment_df[(payment_df['payment_date']>= '2005-01-01') & (payment_df['payment_date']<= '2005-12-31')]
df = df.groupby(['customer_id']).sum()[['amount']]
df[df['amount'] >200].head()
```

```python
# join with customer data set to get names
df[df['amount'] >200].merge(customer_df,on=['customer_id'],how='inner')[['first_name','last_name','amount']]
```

```python
rental2005_df = rental_df[(rental_df['rental_date']>= '2005-01-01') & (rental_df['rental_date']<= '2005-12-31')][['customer_id']]
rental2006_df = rental_df[(rental_df['rental_date']>= '2006-01-01') & (rental_df['rental_date']<= '2006-12-31')][['customer_id']]
df = rental2006_df.merge(rental2005_df, indicator=True, how='outer')
df[df['_merge'] == 'right_only'][['customer_id']].count()

len(df[df['_merge'] == 'right_only']['customer_id'].unique())
```
